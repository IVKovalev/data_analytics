# Market Analysis of Video Games.

**Data and its source.**

The dataset is in a csv file. Open source.

The dataset contains historical data up to 2016 on game sales, user and expert ratings, genres and platforms.

**Description of the data:**
- Name — the name of the game
- Platform — platform
- Year_of_Release — year of release
- Genre — the genre of the game
- NA_sales — sales in North America (millions of copies sold)
- EU_sales — sales in Europe (millions of copies sold)
- JP_sales — sales in Japan (millions of copies sold)
- Other_sales — sales in other countries (millions of copies sold)
- Critic_Score — critics' score (maximum 100)
- User_Score — user rating (maximum 10)
- Rating — rating from the ESRB organization (Entertainment Software Rating Board). This association determines the rating of computer games and assigns them the appropriate age category.


**Objective:**

To conduct a study for an online computer game store in order to identify patterns that determine the success of the game in order to identify a potentially popular product and plan an advertising campaign.
Create a portrait of the user of each region.

**Hypotheses:**
1. The average user ratings of the Xbox One and PC platforms are the same;
2. The average user ratings of the Action and Sports genres are different.


**Analytical research plan:**
 1. Overview of the data.
 2. Data preprocessing.
 3. Data enrichment.
 4. Research data analysis.
 5. Hypothesis testing.
 6. Conclusion.

## Data preprocessing.

### Improving style and readability

```python
# Let's change the name of the columns, reduce them to lowercase and remove unnecessary spaces if any
df.columns
```

```python
# There are no extra spaces, you just need to bring the names to the lower case, and in all columns.
df.columns = df.columns.str.lower()
```

```python
df.head()
```

### Converting data types 

```python
# Let's change the data type in the user_score column to float.
# In the object format, this data will not be available for numerological research.
# To do this, replace the value "tbd" in this column with the value "NaN".
# Because in the sense of "tbd" is the data skipped during filling (to be determined), i.e. "0".
df['user_score'] = df['user_score'].replace('tbd', 'NaN')
df['user_score'] = df['user_score'].astype('float64')
```

```python
# Let's check the resulting data types:
df.info()
```

Now the types look acceptable:

- Text info. in the name, platform, and genre columns, as well as the letter designation of the rating in rating is the *object* type.
- Year - *int* type (we will convert it at the stage of removing omissions).
- Other columns - sales and estimates data - *float* type.

### Filling in missing values

```python
# Once again, let's highlight the missing values in which columns.
# Let's create a "technical" variable "a" for convenience. 
a = pd.DataFrame(df.isna().sum())
a.columns = ['is_null']
a.query('is_null > 0')
```

```python
# Let's look at the lines with a gap in the name column.
df.loc[df['name'].isna()]
```

The name of the game is important information for further conclusions, besides, you can notice that the genre and information about ratings are also missing in the lines. Given that there are only 2 such lines, you can delete them. 

```python
#delete the rows with missing values in the 'name' column. The same lines contain missing values of the 'genre' column
df = df.dropna(subset=['name']).reset_index(drop=True)
```

```python
# Before filling in the missing values in the critic and user score data, let's check for outliers.
# We'll create box plots for `critic_score` and `user_score`.
df.boxplot(column='critic_score', figsize = (5,4))
plt.show()
df.boxplot(column='user_score', figsize=(5,4))
plt.show()
```

Outliers towards decreasing values are visible in both columns. Since the assessment of users and critics is subjective, we will not fill in the gaps with either the average or the median value. Fill in the gaps with the value "0".

```python
df['critic_score'] = df['critic_score'].fillna(0)
df['user_score'] = df['user_score'].fillna(0)
```

```python
# Where there is no data for the entire platform, fill in the gaps with the median value for all platforms.
df['critic_score'] = df['critic_score'].fillna(df['critic_score'].median())
df['user_score'] = df['user_score'].fillna(df['user_score'].median())
```

```python
# Fill in the missing values in the `rating` column with "No data" ("Нет данных").
# The missing values may be due to the rating being assigned by an organization other than ESRB or because the data was not available.

df['rating'] = df['rating'].fillna('Нет данных')
```

```python
# Fill in the missing values in the `year_of_release` column where possible.
# Selectively check 2 games with missing year values.
df[(df['name'] == 'FIFA Soccer 2004') | (df['name'] == 'Inversion')]
```

As you can see, the games were released on different platforms in the same year. It can be assumed that the missing values correspond to the year of release on other platforms.

```python
# Fill in the gaps with the median value for the year of release on other platforms
for i in df['name'].unique():
    df.loc[(df['name'] == i) & (df['year_of_release'].isna()), 'year_of_release'] = \
    df.loc[(df['name'] == i), 'year_of_release'].median()
```

```python
# Additionally, we will check for the absence of the missing values:
df.isna().sum()
```

The remaining missing values in 'year_of_release' are where the game is unique or where the values are missing across all platforms.

```python
# Fill in the remaining gaps with the value "0" and convert the year_of_release type to the int type, as indicated earlier.
df['year_of_release'] = df['year_of_release'].fillna(0).astype('int64')
```

The largest number of missing values was in the evaluation and rating data. In the future, it is advisable to add information about the rating assigned by other agencies.

### The impact on sales of user reviews and critics.

```python
# Let's see how the reviews of users and critics affect sales within one popular platform.
# Let's take the PS3 platform for this. Let's build a scatter plot and calculate the correlation between reviews and sales.
df_a.loc[(df_a['platform'] == 'PS3') & (df_a['critic_score'] > 0)].plot(x = 'critic_score', y = 'total_sales', kind='scatter', alpha = 0.5)
plt.show()
df_a.loc[(df_a['platform'] == 'PS3') & (df_a['user_score'] > 0)].plot(x = 'user_score', y = 'total_sales', kind='scatter',alpha = 0.5)
plt.show()
```

```python
def cor (value):
    return df_a.loc[(df_a['platform'] == 'PS3') & (df_a[value] > 0)][value].corr(df_a.loc[(df_a['platform'] == 'PS3') & (df_a[value] > 0)]['total_sales'])
values = ['critic_score', 'user_score']
for i in values:
    print('The correlation for', i, 'c total_sales is', cor(i))
```

Despite the fact that visually the number of sales is higher for games with a higher rating, the calculation of the Pearson correlation coefficient indicates a low dependence of total sales on the rating of the game.
- The correlation of critics' ratings with total sales is 0.44. **There is a correlation, but it is low**.
- Correlation of user ratings with total sales - -0.016. **There is almost no correlation**. This is probably due to the high sales of individual low-rated games.

Let's check whether the low dependence of total sales on reviews from critics and users is typical for other platforms from the TOP 10 in sales.

```python
for p in top10_sales_platform['platform'].unique():
    def cor (value):
        return df_a.loc[(df_a['platform'] == p) & (df_a[value] > 0)][value].corr(df_a.loc[(df_a['platform'] == p) & (df_a[value] > 0)]['total_sales'])
    values = ['critic_score', 'user_score']
    print('Platform:\033[07m {} \033[0m'.format(p))
    for i in values:
        print('The correlation for', i, 'c total_sales is', cor(i))
```

As you can see above, the situation is typical for all platforms from the TOP 10 in terms of sales:

**Sales are independent of user ratings and slightly dependent on critics' ratings**.

## Hypothesis testing.

### Hypothesis: The average user ratings of the Xbox One and PC platforms are the same.

Let's formulate the null and alternative hypotheses and define the 'alpha' parameter.

- H_0: Average user ratings (user_score) of the XOne platform = average user ratings (user_score) of the PC platform.
- H_a: Average user ratings (user_score) of the XOne platform ≠ average user ratings (user_score) of the PC platform.

alpha = 0.05

```python
a = df[df['platform'] =='XOne']['user_score']
b = df[df['platform'] =='PC']['user_score']

# results = calling a method to test a hypothesis
results = st.ttest_ind(a, b, equal_var = False)
# alpha = setting the significance level value
alpha = .05
# displaying the p-value on the screen
print(results.pvalue)
if results.pvalue < alpha:
    print('We reject the null hypothesis')
else:
    print('It was not possible to reject the null hypothesis')
```

The hypothesis that the average user ratings of the Xbox One and PC platforms are the same turned out to be rejected.

### Hypothesis: The average user ratings of the Action and Sports genres are different.

Let's formulate the null and alternative hypotheses and define the 'alpha' parameter.

- H_0: Average ratings (user_score) of the Action genre ≠ Average ratings (user_score) of the Sports genre.
- H_a: Average ratings (user_score) of the Action genre = Average ratings (user_score) of the Sports genre.

alpha = 0.05

```python
a = df.loc[(df['genre'] =='Action') & (df['user_score'] > 0)]['user_score']
b = df[(df['genre'] =='Sports') & (df['user_score'] > 0)]['user_score']

results = st.ttest_ind(a, b, equal_var = False)
alpha = .05
print(results.pvalue)
if results.pvalue > alpha:
    print('We reject the null hypothesis')
else:
    print('It was not possible to reject the null hypothesis')
```

The hypothesis that the average user ratings of the Action and Sports genres are different turned out to be rejected.

To test the hypothesis of equality of the means of two populations, we applied the **Student's t-test** for independent samples using the **st.ttest_ind** method. Both populations are quite large; however, since their sizes may differ and we can assume that the variances of the populations are different, we specified the equal_var = False parameter.
